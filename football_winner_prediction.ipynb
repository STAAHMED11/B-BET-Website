```python
winner=[]
diffgoals=[]
for i in range (df.Score.count()):
    if (df.iloc[i,36]>df.iloc[i,38]):
        winner.append(1)
    elif(df.iloc[i,36]<df.iloc[i,38]):
        winner.append(-1)
    else :
        winner.append(0)
    diffgoals.append((df.iloc[i,36])-(df.iloc[i,38]))
df['winner']=winner
df['difference of goals']=diffgoals

```

```python
L=['english','french','german','italian','spanish']
Y=[2014,2015,2016,2017,2018,2019,2020]
D={}
for l in L:
    for y in Y:
        D[(l,y)]=df.loc[df['year']==y].loc[df['league']==l]
```

```python
#how the result of last 5 matches can effect the 6th
#this cell will calculate the probability of winning or defeat or null depending on the results of 5 previous games


BL={}
BR={}

for a in range(-1,2):
    for b in range(-1,2):
        for c in range(-1,2):
            for d in range(-1,2):
                for e in range(-1,2):
                    for f in range(-1,2):
                        BL[(a,b,c,d,e,f)]=0
                        BR[(a,b,c,d,e,f)]=0
                
                
for z in PR.keys():
    for i in range (1,len(PR[z])-4):
        for a in range(-1,2):
            for b in range(-1,2):
                for c in range(-1,2):
                    for d in range(-1,2):
                        for e in range(-1,2):
                            for f in range(-1,2):
                                if ((PR[z][i-1]==a) and (PR[z][i]==b) and PR[z][i+1]==c and PR[z][i+2]==d and PR[z][i+3]==e and PR[z][i+4]==f):
                                    BL[(a,b,c,d,e,f)]+=1

                
for a in range(-1,2):
    for b in range(-1,2):
        for c in range(-1,2):
            for d in range(-1,2):
                for e in range(-1,2):
                    for f in range(-1,2):
                        BR[(a,b,c,d,e,f)]=(BL[(a,b,c,d,e,f)]/(BL[(a,b,c,d,e,1)]+BL[(a,b,c,d,e,0)]+BL[(a,b,c,d,e,-1)]))
                   
print((-1,-1,-1,-1,-1,-1),BR[(-1,-1,-1,-1,-1,-1)])
print((1,1,1,1,1,1),BR[(1,1,1,1,1,1)])
print((1,1,1,1,-1,1),BR[(1,1,1,1,-1,1)])
```

```python
P={}
HW={}
AW={}
HP={}
AP={}
PM={}
BS={}
BC={}
for l in L:
    for y in Y:
        
        K=list(D[(l,y)].groupby('Home Team').groups.keys())
        D[(l,y)]['Home Team points bm']=0
        D[(l,y)]['Away Team points bm']=0
        D[(l,y)]['difference of points bm']=0
        D[(l,y)]['Home Team playing winning ']=0
        D[(l,y)]['Away Team playing winning ']=0
        D[(l,y)]['Home Team total goals scored bm']=0
        D[(l,y)]['Away Team total goals scored bm']=0
        D[(l,y)]['Home Team total goals conceeded bm']=0
        D[(l,y)]['Away Team total goals conceeded bm']=0
        D[(l,y)]['Home Team difference of goals bm']=0
        D[(l,y)]['Away Team difference of goals bm']=0 
        
        for k in K:
            P[(l,y,k)]=0
            AW[(l,y,k)]=0
            HW[(l,y,k)]=0
            AP[(l,y,k)]=0
            HP[(l,y,k)]=0
            BS[(l,y,k)]=0
            BC[(l,y,k)]=0
 
        for i in range (D[(l,y)].shape[0]):
            
            #team points
            D[(l,y)].iloc[i,44]=P[(l,y,(D[(l,y)].iloc[i,1]))]
            D[(l,y)].iloc[i,45]=P[(l,y,(D[(l,y)].iloc[i,2]))]
            
            #difference of points
            D[(l,y)].iloc[i,46]=P[(l,y,(D[(l,y)].iloc[i,1]))]-P[(l,y,(D[(l,y)].iloc[i,2]))]
            #home and away winning 
            D[(l,y)].iloc[i,47]=HW[(l,y,(D[(l,y)].iloc[i,1]))]
            D[(l,y)].iloc[i,48]=AW[(l,y,(D[(l,y)].iloc[i,2]))]
            
            #calcul of team points
            if ((D[(l,y)].iloc[i,42])==1):
                P[(l,y,(D[(l,y)].iloc[i,1]))]+=3
            elif ((D[(l,y)].iloc[i,42])==-1):
                P[(l,y,(D[(l,y)].iloc[i,2]))]+=3
            elif ((D[(l,y)].iloc[i,42])==0):
                P[(l,y,(D[(l,y)].iloc[i,2]))]+=1
                P[(l,y,(D[(l,y)].iloc[i,1]))]+=1
            #calcul of number of wins of playing home for the home team
            if (D[(l,y)].iloc[i,42]==1):
                HW[(l,y,(D[(l,y)].iloc[i,1]))]=(HW[(l,y,(D[(l,y)].iloc[i,1]))]+1)
            #calcul of number of wins of playing away for the away team
            if (D[(l,y)].iloc[i,42]==-1):
                AW[(l,y,(D[(l,y)].iloc[i,2]))]=(AW[(l,y,(D[(l,y)].iloc[i,2]))]+1)
                
                
            #calcul bilan of  goals before match    
            D[(l,y)].iloc[i,49]=BS[(l,y,(D[(l,y)].iloc[i,1]))]
            D[(l,y)].iloc[i,50]=BS[(l,y,(D[(l,y)].iloc[i,2]))]
            D[(l,y)].iloc[i,51]=BC[(l,y,(D[(l,y)].iloc[i,1]))]
            D[(l,y)].iloc[i,52]=BC[(l,y,(D[(l,y)].iloc[i,2]))]
            
            D[(l,y)].iloc[i,53]=D[(l,y)].iloc[i,49]-D[(l,y)].iloc[i,51]
            D[(l,y)].iloc[i,54]=D[(l,y)].iloc[i,50]-D[(l,y)].iloc[i,53]
            BS[(l,y,(D[(l,y)].iloc[i,1]))]+=D[(l,y)].iloc[i,36]
            BS[(l,y,(D[(l,y)].iloc[i,2]))]+=D[(l,y)].iloc[i,37]
            BC[(l,y,(D[(l,y)].iloc[i,1]))]+=D[(l,y)].iloc[i,38]
            BC[(l,y,(D[(l,y)].iloc[i,2]))]+=D[(l,y)].iloc[i,39]
```

```python
def proba_bar(HT,AT):
  x_data= [list(list(X_test.loc[X_test['Home Team']==HT].loc[X_test['Away Team']==AT]['pred'])[0])]
  top_labels = ['Home','Draw','Away']
  colors = ['rgba(38, 24, 74, 0.8)', 'rgba(71, 58, 131, 0.8)','rgba(122, 120, 168, 0.8)']
  y_data = ['proba']
  fig = go.Figure()

  for i in range(0, len(x_data[0])):
    for xd, yd in zip(x_data, y_data):
      fig.add_trace(go.Bar(
            x=[xd[i]], y=[yd],
            orientation='h',
            marker=dict(color=colors[i],line=dict(color='rgb(248, 248, 249)', width=1))))

  annotations = []
  for yd, xd in zip(y_data, x_data):
    annotations.append(dict(xref='x', yref='y',x=xd[0] / 2, y=yd,text=str(xd[0])[:5] + '%',font=dict(family='Arial', size=25,color='rgb(248, 248, 255)'),showarrow=False))
    if yd == y_data[-1]:
      annotations.append(dict(xref='x', yref='paper',x=xd[0] / 2, y=1.1,text=top_labels[0],font=dict(family='Arial', size=25,color='rgb(67, 67, 67)'), showarrow=False))
      space = xd[0]
  for i in range(1, len(xd)):
    annotations.append(dict(xref='x', yref='y',x=space + (xd[i]/2), y=yd,text=str(xd[i])[:5] + '%',font=dict(family='Arial', size=25,color='rgb(248, 248, 255)'),showarrow=False))
    if yd == y_data[-1]:
      annotations.append(dict(xref='x', yref='paper',x=space + (xd[i]/2), y=1.1,text=top_labels[i],font=dict(family='Arial', size=25,color='rgb(67, 67, 67)'),showarrow=False))
      space += xd[i]

  fig.update_layout(annotations=annotations)
  fig.update_layout(
    barmode='stack',
    paper_bgcolor='rgb(248, 248, 255)',
    plot_bgcolor='rgb(248, 248, 255)',
    margin=dict(l=120, r=10, t=140, b=80),
    showlegend=False)
  return fig
proba_bar('BAYERN','DORTMUND')
```

```python
def results(HT1,AT1,league):
    ind=D_league[league].loc[(D_league[league]['Home Team']==HT1) & (D_league[league]['Away Team']==AT1)].index[0]
    HR=PR[(league,2020,HT1)][:ind//10]
    AR=PR[(league,2020,AT1)][:ind//10]
    result1=pd.DataFrame([[0,0,0]]*(ind//10),columns=['index','Team','result'])
    result2=pd.DataFrame([[0,0,0]]*(ind//10),columns=['index','Team','result'])
    for i in range(ind//10):
      result1.iloc[i,1:]=[HT1,HR[i]]
      result1.iloc[i,0]=i+1

    for j in range(ind//10):
      result2.iloc[j,1:]=[AT1,AR[j]]
      result2.iloc[j,0]=j+1

    result=pd.concat([result1,result2])
    for i in range(result.shape[0]):
      if result.iloc[i,-1]==1:
        result.iloc[i,-1]='win'
      if result.iloc[i,-1]==0:
        result.iloc[i,-1]='draw'
      if result.iloc[i,-1]==-1:
        result.iloc[i,-1]='lose'
    fig = px.scatter(result, x="index", y="Team", color='result',size=[60]*2*(ind//10))

    return fig
results('PSG','DIJON','french')
```

```python
def index1 (HT,AT,league):
    ind1=D_league[league].loc[(D_league[league]['Home Team']==HT) & (D_league[league]['Away Team']==AT)].index[0]
    return ind1
index1('CHELSEA','FULHAM','english')
```

```python
def Bar (HT,AT,C,league):

  ind=index1(HT,AT,league)
  H=[]
  A=[]
  for c in C:
    val=D_league[league].loc[D_league[league]['Home Team']==HT].iloc[:1+ind//10]['Home Team '+c].sum()/D_league[league].loc[D_league[league]['Home Team']==HT].iloc[:1+ind//10]['Home Team '+c].count()
    H+=[[HT,c,val]]
  for c in C:
    val=D_league[league].loc[D_league[league]['Away Team']==AT].iloc[:1+ind//10]['Away Team ' +c].sum()/D_league[league].loc[D_league[league]['Away Team']==AT].iloc[:1+ind//10]['Away Team '+c].count()
    A+=[[AT,c,val]]
  VIS=H+A
  DC1=pd.DataFrame(VIS,columns=['Team','target','value'])
  fig = px.histogram(DC1, x="target", y="value", color="Team", text_auto=True, barmode='group',
             histfunc='avg')
  return fig
Bar('CHELSEA','FULHAM',['Total Shots','On Target Shots','Goals Scored'],'english')
```

```python
def Line(HT,AT,c,league) :
  line0=pd.DataFrame([[0,0]],columns=['Team','values'])
  line0=line0.drop(0)
  line1=pd.DataFrame([[0,0]],columns=['Team','values'])
  line1=line1.drop(0)
  ind=D_league[league].loc[(D_league[league]['Home Team']==HT) & (D_league[league]['Away Team']==AT)].index[0]
  line0['values']=D_league[league].loc[D_league[league]['Home Team']==HT].iloc[:1+ind//10]['Home Team '+c]
  line1['values']=D_league[league].loc[D_league[league]['Away Team']==AT].iloc[:1+ind//10]['Away Team '+c]
  for i in range(ind):
    line0['Team']=HT
  for i in range(ind):
    line1['Team']=AT
  line0=line0.reset_index()
  line1=line1.reset_index()
  line0=line0.drop('index',axis=1)
  line1=line1.drop('index',axis=1)
  line0=line0.reset_index()
  line1=line1.reset_index()
  line=pd.concat([line0,line1])
  fig = px.line(line, x="index", y="values", color='Team')
  return fig
```

```python
import plotly.subplots as sp
from google.colab import files

def subplot(HT,AT,league):
  Figures=[results(HT,AT,league),
           Bar(HT,AT,['Total Shots','On Target Shots','Goals Scored'],league),
           Bar(HT,AT,['Possession %','Pass Success %'],league),
           Bar(HT,AT,['Clearances','Throw Ins','Aerials Won'],league),
           Bar(HT,AT,['playing winning '],league),
           Line(HT,AT,'points bm',league),
           Line(HT,AT,'total goals scored bm',league),
           Line(HT,AT,'mean rating bm',league),
           Line(HT,AT,'accuracy shots bm',league)]
  titles=('previous results','Total Shots                              On Target Shots                                Goals Scored',
          'Possession %                                    Pass Success %',
          'Clearances                                  Throw Ins                                      Aerials Won',
        'playing winning ','points bm','total goals scored bm','mean rating bm','accuracy shots bm')
  Figures[0].update_layout(
    barmode='stack',
    paper_bgcolor='rgb(248, 248, 255)',
    plot_bgcolor='rgb(248, 248, 255)',
    margin=dict(l=120, r=10, t=140, b=80),
    showlegend=False)
  figure_traces=[]
  for i in range(len(Figures)):
    figure_traces+=[[]]
  r=0
  for figure in Figures:
    for trace in range(len(figure["data"])):
      figure_traces[r].append(figure["data"][trace])
    r+=1
  this_figure = sp.make_subplots(rows=10, cols=1,subplot_titles=titles)
  j=1
  for figure in Figures:
    for traces in figure_traces[j-1]:
      this_figure.append_trace(traces, row=j, col=1)
    j+=1
  this_figure.update_layout(title_text=HT+'  Vs  '+AT)
  return this_figure
```
